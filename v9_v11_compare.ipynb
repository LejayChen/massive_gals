```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import os
from astropy.table import *
from tqdm import tqdm
import matplotlib
plt.rc('font', family='serif'), plt.rc('xtick', labelsize=18), plt.rc('ytick', labelsize=18)
plt.rcParams['savefig.dpi'] = 300
plt.rc('text',usetex=True)
matplotlib.rcParams['xtick.direction'] = 'in'
matplotlib.rcParams['ytick.direction'] = 'in'
np.set_printoptions(precision=3)
```

```python
# load v9 catalog
cat_v9 = Table.read('clauds_cat_v2020/completeness_useful_params_cat/COSMOS_deep_v9_gal_cut_params_sfq_added.fits')
cat_v9 = cat_v9[cat_v9['inside_u'] == True]
cat_v9 = cat_v9[cat_v9['MASK'] == 0]  # unmasked
cat_v9_gal = cat_v9[cat_v9['OBJ_TYPE'] == 0]  # galaxies

# load v11 catalog
cat_v11 = Table.read('/Volumes/LejaySSD/v11_cats/COSMOS_v11_11bands_v211224_cut_params.fits')
cat_v11 = cat_v11[cat_v11['inside_u'] == True]
cat_v11 = cat_v11[cat_v11['MASK'] == 0]  # unmasked
cat_v11_gal = cat_v11[cat_v11['OBJ_TYPE'] == 0]  # galaxies

print(len(cat_v9_gal),len(cat_v11_gal))
```

```python
# z distribution
fig, axs = plt.subplots(1,1, figsize = (18, 7),sharex=True)

plt.hist(cat_v9_gal['ZPHOT'],range=(0,6),bins='scott',histtype='step',density=True)
plt.hist(cat_v11_gal['ZPHOT'],range=(0,6),bins='scott',histtype='step',density=True)
plt.xlabel('$z$',fontsize=18)
```

```python
# mass comparison for matched catalog
fig, axs = plt.subplots(1,1, figsize = (18, 7),sharex=True)

plt.scatter(cat_v9_gal['MASS_MED'],cat_v11_gal['MASS_MED'],s=1,color='k')
plt.xlim([7,13])
plt.ylim([7,13])
plt.xlabel('$M_{v9}$',fontsize=18)
plt.xlabel('$M_{v11}$',fontsize=18)
```
