```python
import pymc3 as pm
from astropy.table import Table
import matplotlib.pyplot as plt

plt.rc('text',usetex=True)
plt.rc('font', family='serif'), plt.rc('xtick', labelsize=15), plt.rc('ytick', labelsize=15)
```

```python
cat = Table.read('CUT_deep_catalogs/CUT3_DEEP_deep.fits')
cat_gal = cat[cat['preds_median']<0.89]
cat_gal_z_slice = cat_gal[abs(cat_gal['zKDEPeak']-0.6)<0.1]
```

```python
mass_list = cat_gal_z_slice['MASS_MED']
ssfr_list = cat_gal_z_slice['SSFR_BEST']
sfprob_list = cat_gal_z_slice['sfProb']
```

```python
fig = plt.figure(figsize=(9,9))
plt.scatter(mass_list,ssfr_list,marker='.',color='k',alpha=0.1)
plt.xlim([6.8,12])
plt.ylim([-15,-6])
plt.xlabel('$\log(M/M_\odot)$',fontsize=16)
plt.ylabel('$\log(\mathrm{SSFR})$ [$\log($yr$^{-1}$)]',fontsize=16)
plt.savefig('figures/mass_ssfr.png',dpi=200)
plt.show()
```

```python
fig = plt.figure(figsize=(9,9))
plt.scatter(mass_list,sfprob_list,marker='.',color='k',alpha=0.2)
plt.xlim([6.8,12])
plt.ylim([0,1])
plt.xlabel('$\log(M/M_\odot)$',fontsize=16)
plt.ylabel('sf-Prob',fontsize=16)
plt.savefig('figures/mass_sfprob.png',dpi=200)
plt.show()
```
